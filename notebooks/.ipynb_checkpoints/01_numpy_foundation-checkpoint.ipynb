```python
A = np.array([
    [3, 6, 9],
    [2, 4, 8],
    [5, 7, 1]
])

print(A[1, 2]) # 8
print(A[:, 0]) # [3 2 5]
print(A[0:2, 1:3]) # [[6 9] [4 8]]
```

```python
A > 4
```

```python
A[A > 4]
```

```python
# AND condition
A[(A > 3) & (A < 8)]
```

```python
# OR condition
A[(A < 3) | (A > 7)]
```

```python
# Conditional Replacement (Feature Cleaning)
# This is extremely common in preprocessing.
A[A < 5] = 0
A
```

```python
# Real ML Use Case Pattern
# Example: Clipping noisy values
A[A > 7] = 7
A[A < 3] = 3
A
```

```python
# Masking with Copy Safety
# If you want to preserve original data:
B = A.copy()
B[B > 5] = -1
B
```

```python
A = np.array([
    [12, 5, 7],
    [3, 9, 2],
    [8, 6, 4]
])

# Tasks:
# Extract all values divisible by 2
# Replace all values > 6 with -1
# Extract values between 4 and 9 inclusive
```

```python
# 1. Extract all values divisible by 2
print(A[A % 2 == 0])
```

```python
# 2. Replace all values > 6 with -1
A[A > 6] = -1
print(A)
```

```python
# 3. Extract all values between 4 and 9 incluisve
A[(A > 3) & (A < 10)]
```

```python
# Broadcasting in Numpy
A = np.array([
    [2, 4, 6],
    [1, 3, 5]
])

B = np.array([10, 20, 30])
```

## Array Creation & Shape Manipulation

```python
# Building arrays without manual data
zeros = np.zeros((3, 4))       # placeholder feature matrix
ones = np.ones((2, 3))         # bias initialization
identity = np.eye(3)           # identity matrix (used in linear algebra)
full = np.full((2, 2), 7.0)    # constant fill

print("zeros:\n", zeros)
print("ones:\n", ones)
print("identity:\n", identity)
print("full:\n", full)
```

```python
# Ranges and evenly spaced values
steps = np.arange(0, 10, 2)         # [0, 2, 4, 6, 8]
linear = np.linspace(0, 1, 5)       # 5 points between 0 and 1 (normalization scale)

print("arange:", steps)
print("linspace:", linear)
```

```python
# Random arrays — you'll use these constantly for synthetic data and weight init
rng = np.random.default_rng(seed=42)   # reproducible (always set a seed in ML)

rand_uniform = rng.random((3, 3))       # values between 0 and 1
rand_normal = rng.standard_normal((3, 3))  # Gaussian — what neural net weights use
rand_int = rng.integers(0, 100, size=(4, 5))  # random integers

print("uniform:\n", rand_uniform.round(3))
print("\nnormal:\n", rand_normal.round(3))
print("\nintegers:\n", rand_int)
```

## Reshape, Flatten, Squeeze

```python
a = np.arange(12)           # [0, 1, 2, ..., 11]
print("original:", a.shape)  # (12,)

matrix = a.reshape(3, 4)    # 3 rows, 4 cols
print("reshaped:\n", matrix)
print("shape:", matrix.shape)

# -1 means "figure it out for me"
auto = a.reshape(2, -1)     # 2 rows, numpy calculates cols=6
print("\nreshape(2,-1):\n", auto)
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])

flat = matrix.flatten()     # always returns a copy
ravel = matrix.ravel()      # returns a view (faster, careful with mutation)

print("flatten:", flat)
print("ravel:", ravel)

# Common ML pattern: flatten image pixels before feeding to a model
image = np.random.default_rng(0).integers(0, 256, size=(28, 28))  # fake grayscale image
pixels = image.flatten()
print("\nimage shape:", image.shape, "→ flattened:", pixels.shape)
```

```python
# expand_dims — adds a dimension (needed when sklearn expects 2D input)
single_sample = np.array([1.5, 2.3, 0.8, 4.1])   # one row of features
print("original shape:", single_sample.shape)       # (4,) — 1D, sklearn will reject this

reshaped = single_sample.reshape(1, -1)             # (1, 4) — 1 sample, 4 features
print("reshaped for sklearn:", reshaped.shape)
```
